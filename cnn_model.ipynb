```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import re
```

```python
dir_list=os.listdir('..')[:11]
dir_list.remove("fase11")
```

```python
current=".."
total=pd.DataFrame(columns=[["PGV1","PGV2","PGV3","ATB1","ATB2","fase","OP"]])
total['PGV1']=total['PGV1'].astype("float64")
total['PGV2']=total['PGV2'].astype("float64")
total['PGV3']=total['PGV3'].astype("float64")

total['ATB1']=total['ATB1'].astype("float64")
total['ATB2']=total['ATB2'].astype("float64")

total['fase'] = total['fase'].astype(int)

test=pd.DataFrame(columns=[["PGV1","PGV2","PGV3","ATB1","ATB2","fase","OP"]])
test['PGV1']=test['PGV1'].astype("float64")
test['PGV2']=test['PGV2'].astype("float64")
test['PGV3']=test['PGV3'].astype("float64")

test['ATB1']=test['ATB1'].astype("float64")
test['ATB2']=test['ATB2'].astype("float64")



test['fase'] = test['fase'].astype(int)
i=0
for path, subdirs, files in os.walk(current):
    #print(subdirs)
    for name in files:
        if 'ref' in path or name[0:1] == ".":
            continue
        names=("BEP" in name or "DPL" in name or "PL" in name or "HL" in name or "FL" in name)
        heads=("h12" in name)# or "h18" in name or "h24" in name or "h30" in name)
        if name.endswith(".csv") and heads and names and "Properties" not in path and "fase11" not in path:
            name_=os.path.join(path, name)
            print(name_)
            df_read = pd.read_csv(name_)
            df=pd.DataFrame(df_read["PGV1"][:200000])
            df["PGV2"]=pd.DataFrame(df_read["PGV2"][:200000])
            df["PGV3"]=pd.DataFrame(df_read["PGV3"][:200000])
            
            df["ATB1"]=pd.DataFrame(df_read["ATB1"][:200000])
            df["ATB2"]=pd.DataFrame(df_read["ATB2"][:200000])
            
            df1=pd.DataFrame(df_read["PGV1"][200000:200000+40000])
            df1["PGV2"]=pd.DataFrame(df_read["PGV2"][200000:200000+40000])
            df1["PGV3"]=pd.DataFrame(df_read["PGV3"][200000:200000+40000])
            
            df1["ATB1"]=pd.DataFrame(df_read["ATB1"][200000:200000+40000])
            df1["ATB2"]=pd.DataFrame(df_read["ATB2"][200000:200000+40000])
            
            if "BEP" in name:
                #print("BEP")
                df["OP"]=str("BEP")
                df1["OP"]=str("BEP")
            if "DPL" in name:
                df["OP"]=str("DPL")
                df1["OP"]=str("DPL")
                                    

            
            x1 = np.linspace(0, len(df), len(df), endpoint = False)
            for fase in dir_list:
                if fase+"/" in path:
                    num=re.findall(r'\d+', fase)
                    df["fase"]=int(num[0])
                    df['fase'] = df['fase'].astype(int)
                    df1["fase"]=int(num[0])
                    df1['fase'] = df1['fase'].astype(int)
                    
                    
                    if total.empty:
                        total=df
                    if test.empty:
                        test=df1
                        
                    else:
                    #print(df,"--")
                    #total.loc[len(total.index)]=
                        #pd.merge([total,df],on = ['PGV1', 'fase'])
                        total=pd.concat([total,df])
                        test=pd.concat([test,df1])
#                     #df.plot()
#                     plt.scatter(df,range(len(df)), color = colors[i])
                    #plt.scatter(x1,df["PGV1"], color = colors[i],label=fase)
                    
                            
            i+=1
#plt.legend()
#plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Assuming you have your time series data stored in 'X_train' and corresponding labels in 'y_train'
from sklearn.preprocessing import MinMaxScaler
optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-8
)
# Assuming you have your time series data stored in 'X_train' and corresponding labels in 'y_train'

# Normalize the time series data
# scaler = MinMaxScaler()
# x_train_normalized = scaler.fit_transform(x_train)

model = models.Sequential()
model.add(layers.Conv1D(128, 8, activation='relu', input_shape=(20, 5)))
#model.add(layers.SpatialDropout1D(0.2))
model.add(layers.BatchNormalization())
#model.add(layers.MaxPooling1D(2))

model.add(layers.Conv1D(256, 5, activation='relu'))
#model.add(layers.SpatialDropout1D(0.3))
model.add(layers.BatchNormalization())
#model.add(layers.MaxPooling1D(2))

model.add(layers.Conv1D(128, 3, activation='relu'))
#model.add(layers.SpatialDropout1D(0.3))
model.add(layers.BatchNormalization())
#model.add(layers.MaxPooling1D(2))

#model.add(layers.Flatten())
model.add(layers.GlobalAveragePooling1D())


#model.add(layers.Dense(64, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.1)))
model.add(layers.Dense(10, activation='softmax')) # Replace 'num_classes' with the number of output classes

# Compile the model
model.compile(optimizer=optimizer,loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])

# Reshape your training data to match the input shape of the model
#x_train = x_train.reshape((x_train.shape[0], 915, 1))
#x_train_normalized = x_train_normalized.reshape((x_train_normalized.shape[0], 915, 1))
# Train the model
model.fit(x_train_n, tf.keras.utils.to_categorical(y_train), epochs=10, batch_size=16)

# Assuming you have your test data stored in 'X_test' and corresponding labels in 'y_test'

# Reshape your test data to match the input shape of the model
#x_test = x_test.reshape((x_test.shape[0], 915, 1))
#x_test_normalized = scaler.transform(x_test)

# Reshape your normalized test data to match the input shape of the model
#x_test_normalized = x_test_normalized.reshape((x_test_normalized.shape[0], 915, 1))
# Evaluate the model
test_loss, test_acc = model.evaluate(x_test_n, tf.keras.utils.to_categorical(y_test), verbose=2)
print("Test accuracy:", test_acc)
```
